This is an implementation of the Linear Regression Model using the scikit-learn module.
The model is has an approximate efficiency of 40% which is not enough.
More refinements need to be made.

```python
# Loeading the required libraries
# Modules used : Pandas, Scikit-Learn, Numpy
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# Loading the training data set
df = pd.read_csv("bpp_training_data.csv")
df # overview of the data set
```

```python
# Removing potentially unwanted features
del df['SalesID']
del df['MachineID']
del df['ModelID']
del df['datasource']
del df['auctioneerID']
df
```

```python
# Creating a new DataFrame with only the target variable
df_y = df[['SalePrice']]

# Deleting the target variable column from the original DataFrame
del df['SalePrice']
```

```python
# Creating a linear regression object
reg = linear_model.LinearRegression()
```

```python
# Splitting 20% of the Data into test set. Will be used later to evaluate the model
x_train, x_test, y_train, y_test = train_test_split(df, df_y, test_size=0.2, random_state=4)
```

```python
# Fitting the linear model to the training data
reg.fit(x_train, y_train)
```

```python
# The coefficients of the Linear Regression hypothesis function
reg.coef_
```

```python
# Intercept of the Linear Regression Hypothesis function
reg.intercept_
```

```python
# Making predictions over the test set
a = reg.predict(x_test)
a
```

```python
# Calculating the mean square error
print(np.mean((a - y_test)**2))
```

```python
# Evaluating our model on the test set to know its approximate efficiency
reg.score(x_test, y_test)
```
